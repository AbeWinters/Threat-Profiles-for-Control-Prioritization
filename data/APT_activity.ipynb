# APT Activity Experiment
This notebook contains code to aim to answer the following question: How active are APTs after their first operation?
Advanced APT attack operations will remain several for several years. [APT Cyber-Attack Modeling](https://books.google.nl/books?hl=nl&lr=&id=Shd2EAAAQBAJ&oi=fnd&pg=PA121&dq=apt+cyber+attack+activity&ots=syhFrdDUk7&sig=ejQ0v5LJ12c_SPUZe9f7pt8cmq0#v=onepage&q=apt%20cyber%20attack%20activity&f=false) shows the difference in first detection versus their first operation.

The results are used to choose a weighting function for operations in the actor ranking.

```python
import pandas as pd

df = pd.read_json('ETDA_ATTCK_merge.json')
df = df[['etda_name','etda_first_seen','etda_operation_year']]
df['total_ops'] = df['etda_operation_year'].apply(len)
df.sort_values(by='total_ops')
```

Count the number of operations in the same year as they are first seen

```python
df = df.explode('etda_operation_year')

df['etda_operation_year'] = df['etda_operation_year'].where(df['etda_operation_year'] == df['etda_first_seen'],'')
df.loc[df['etda_operation_year'] != '']
df = df.groupby(['etda_name']).agg({'etda_name':'first','etda_first_seen':'first','total_ops':'first','etda_operation_year':'list'})
df.sort_values(by='etda_operation_year',ascending=False)
```
