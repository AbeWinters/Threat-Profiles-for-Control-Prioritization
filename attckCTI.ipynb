```python
from attackcti import attack_client
from pandas import json_normalize
import sys

def get_attck_dataframe(matrix = 'enterprise'):
    if  (matrix.lower() == 'enterprise'):
        # Instantiating attack_client class
        lift = attack_client()
        # Getting techniques for windows platform - enterprise matrix
        attck = lift.get_enterprise_techniques(stix_format = False)
        # Generating a dataframe with information collected
        attck = json_normalize(attck)
        return attck
    else:
        sys.exit('ERROR: Only Enterprise available!!')

# Get enterprise techniques
attck = get_attck_dataframe()
print('Number of TTPs:',len(attck))

attck.head()
```

```python
# These functions are copied from: https://github.com/Cyb3rPandaH/attack-datasources/blob/main/docs/scripts/notebook_functions.py
# Function to import ATT&CK content as a list of dictionaries

def get_mitigation(matrix = 'enterprise'):
    if  (matrix.lower() == 'enterprise'):
        # Instantiating attack_client class
        lift = attack_client()
        # Getting techniques for windows platform - enterprise matrix
        attck = lift.get_enterprise_mitigations(stix_format = False)
        # Removing revoked techniques
        # attck = lift.remove_revoked(attck)
        return attck
    else:
        sys.exit('ERROR: Only Enterprise available!!')

mitigations = json_normalize(get_mitigation())
print(mitigations.columns)
mitigations
```
