# Experiment with weight functions

## Graphs
Graphs for
- 1/x
- 1/x^2
- 1/sqrt(x)
- 1/log(x)

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(0,10,1)

y1 = 1/(1+x) 
y2 = 1/np.exp2(1+x)
y3 = 1/np.sqrt(1+x)
y4 = np.exp2(-x-1)
K = 2.5
w = K*np.exp(-x)

# plt.plot(x,w, label="2*e^-x")
plt.plot(x,y1,label="5.5a: 1/(1+x)")
plt.plot(x,y3, label="5.5b: 1/sqrt(1+x)")
plt.plot(x,y2,label="5.5c: 2^-(1+x)")
# plt.plot(x,y4, label="1/(1 + x)")
plt.legend()
plt.title("Plots of various inverse functions")
plt.savefig("results/figures/ranking/inverse_functions.pdf",format="pdf", bbox_inches="tight")
plt.show()
```

Before choosing an inverse weighting function for the operations and the newness, requirements need to be set.
- The function should not deteriorate too quickly
- Since the buzz will be multiplied to the weighted operation count, it should be taken into account that in can decrease the result rather fast. 

The importance of newness and recent operations, or the lack of importance of older operations, should be defined. For this, an experiment is set up to answer the following question: How active are APTs generally after they are first seen?

Based on the graph, I decide to pursue the 1/x for the operations weighing since it prioritizes recent operations more compared to the inverse square root. 
For the buzz the inverse square root is chosen, since it does not diminish the number of operations too quickly.
